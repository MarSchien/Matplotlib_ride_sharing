```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
cityRead = pd.read_csv(city_data_to_load)
rideRead = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combinedTable = pd.merge(cityRead, rideRead, on='city', how= 'left')

# Display the data table for preview
combinedTable

```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types


# filter merged table above for the 'type' column values equal to rural
ruralCitiesDf = combinedTable.loc[combinedTable['type'] =='Rural']

# group by city column and calculate mean of each column
ruralAvg = ruralCitiesDf.groupby('city').mean()
# just the series of column 'fare' averages, grouped by city
ruralAvgFare = ruralAvg['fare']

# group by city column and calculate mean of each column
ruralCount = ruralCitiesDf.groupby('city').mean()
# just the series of column 'driver_count' averages, grouped by city <the average of all the same numbers is the same number>
ruralDriverCount = ruralCount['driver_count']

# group by city column and calculate the count of each column
ruralRideCount = ruralCitiesDf.groupby('city').count()
# just the series of column 'ride_id' count, grouped by city
ruralRideCountId = ruralRideCount['ride_id']

#repeat same calculations for other 'type' values

suburbanCitiesDf = combinedTable.loc[combinedTable['type'] =='Suburban']


suburbanAvg = suburbanCitiesDf.groupby('city').mean()
suburbanAvgFare = suburbanAvg['fare']

suburbanCount = suburbanCitiesDf.groupby('city').mean()
suburbanDriverCount = suburbanCount['driver_count']

suburbanRideCount = suburbanCitiesDf.groupby('city').count()
suburbanRideCountId = suburbanRideCount['ride_id']



urbanCitiesDf = combinedTable.loc[combinedTable['type'] =='Urban']


urbanAvg = urbanCitiesDf.groupby('city').mean()
urbanAvgFare = urbanAvg['fare']

urbanCount = urbanCitiesDf.groupby('city').mean()
urbanDriverCount = urbanCount['driver_count']

urbanRideCount = urbanCitiesDf.groupby('city').count()
urbanRideCountId = urbanRideCount['ride_id']


# create scatter plots for each city 'type' based on calcs. Set equal to handles that can be used for legend
rural = plt.scatter(x=ruralRideCountId, y=ruralAvgFare, s=20*ruralDriverCount, color="gold", 
            edgecolors='black', marker='o', alpha=0.45, linewidth=0.6,label="Rural")

suburban = plt.scatter(x=suburbanRideCountId, y=suburbanAvgFare, s=20*suburbanDriverCount, color="lightskyblue", 
            edgecolors='black', marker='o', alpha=0.45, linewidth=0.6,label="Suburban")

urban = plt.scatter(x=urbanRideCountId, y=urbanAvgFare, s=20*urbanDriverCount, color="lightcoral", 
            edgecolors='black', marker='o', alpha=0.45, linewidth=0.6,label="Urban")

# create legend with handles
legend = plt.legend(handles=[rural, suburban, urban], loc="best")

# adjust size of legend dots
legend.legendHandles[0]._sizes = [30]
legend.legendHandles[1]._sizes = [30]
legend.legendHandles[2]._sizes = [30]

# label axis and title
plt.title("Pyber Ride Sharing Data")
plt.xlabel('Total Number of Rides Per City')
plt.ylabel('Average Fare ($)')

# create note on the side of the chart
plt.text(50, 33, "Circle size correlates with driver count per city.", horizontalalignment='left', 
         size='medium', color='black', weight='semibold')

# save fig to a path of choice, in this case the desktop of my computer
plt.savefig("/Users/marlonschieber/Desktop/pyberscatter.png", bbox_inches='tight', pad_inches=2)
```

```python
# Show plot


plt.show()
```

## Total Fares by City Type

```python
# calculate total sum of all fares across all city 'type'
totalFare = combinedTable['fare'].sum()

#filter out rural type column 
ruralTypeOnly = combinedTable.loc[combinedTable['type'] =='Rural']
#calculate sum of 'fare' column
ruralFareSum = ruralTypeOnly['fare'].sum()
#divide rural 'fare' sum by total sum of all fares across all city 'type'
percentRuralFare = (ruralFareSum/totalFare)*100

#repeat calculations for other city 'type'
suburbanTypeOnly = combinedTable.loc[combinedTable['type'] =='Suburban']
suburbanFareSum = suburbanTypeOnly['fare'].sum()
percentSuburbanFare = (suburbanFareSum/totalFare)*100

urbanTypeOnly = combinedTable.loc[combinedTable['type'] =='Urban']
urbanFareSum = urbanTypeOnly['fare'].sum()
percentUrbanFare = (urbanFareSum/totalFare)*100

# create lists of values needed for pie chart creation
percent = [percentUrbanFare, percentSuburbanFare, percentRuralFare ]
labels = ['Urban', 'Suburban', 'Rural']
explode = [0.2,0,0]
colors = ['lightcoral', 'lightskyblue','gold']

# create pie chart
plt.pie(percent, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=165)

# label title of pie chart
plt.title("Percent of Total Fares by City Type:")

# save fig to a path of choice, in this case the desktop of my computer
plt.savefig("/Users/marlonschieber/Desktop/totalfares.png", bbox_inches='tight', pad_inches=2)
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# calculate total count of all 'ride_id' across all city 'type'
totalCount = combinedTable['ride_id'].count()

#filter out rural type column 
ruralTypeOnly = combinedTable.loc[combinedTable['type'] =='Rural']
#calculate count of 'ride_id' column
ruralRideCount = ruralTypeOnly['ride_id'].count()
#divide rural 'ride_id' count by total count of all 'ride_id' across all city 'type'
percentRuralCount = (ruralRideCount/totalCount)*100

#repeat calculations for other city 'type'
suburbanTypeOnly = combinedTable.loc[combinedTable['type'] =='Suburban']
suburbanRideCount = suburbanTypeOnly['ride_id'].count()
percentSuburbanCount = (suburbanRideCount/totalCount)*100

urbanTypeOnly = combinedTable.loc[combinedTable['type'] =='Urban']
urbanRideCount = urbanTypeOnly['ride_id'].count()
percentUrbanCount = (urbanRideCount/totalCount)*100

# create lists of values needed for pie chart creation
percent = [percentUrbanCount, percentSuburbanCount, percentRuralCount ]
labels = ['Urban', 'Suburban', 'Rural']
explode = [0.2,0,0]
colors = ['lightcoral', 'lightskyblue','gold']

# create pie chart
plt.pie(percent, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=165)

# label title of pie chart
plt.title("Percent of Total Rides by City Type:")

# save fig to a path of choice, in this case the desktop of my computer
plt.savefig("/Users/marlonschieber/Desktop/totalrides.png", bbox_inches='tight', pad_inches=2)
```

## Total Drivers by City Type

```python
# drop duplicates of values in column'city'
totalDriverCountDropDuplicates = combinedTable.drop_duplicates('city')
# calculate total sum of all 'driver_count' across all city 'type'
totalDriverCount= totalDriverCountDropDuplicates['driver_count'].sum()

# filter out rural type column
ruralTypeOnly = combinedTable.loc[combinedTable['type'] =='Rural']
# drop duplicates of values in column'city'
ruralDuplicateDrop = ruralTypeOnly.drop_duplicates('city')
# calculate sum of'driver_count' column
ruralDriverCount = ruralDuplicateDrop['driver_count'].sum()
#divide rural 'driver_count' by total sum of all 'driver_count' across all city 'type'
percentRuralDriverCount = (ruralDriverCount/totalDriverCount)*100

# repeat calcs for other city 'type'
suburbanTypeOnly = combinedTable.loc[combinedTable['type'] =='Suburban']
suburbanDuplicateDrop = suburbanTypeOnly.drop_duplicates('city')
suburbanDriverCount = suburbanDuplicateDrop['driver_count'].sum()
percentSuburbanDriverCount = (suburbanDriverCount/totalDriverCount)*100

urbanTypeOnly = combinedTable.loc[combinedTable['type'] =='Urban']
urbanDuplicateDrop = urbanTypeOnly.drop_duplicates('city')
urbanDriverCount = urbanDuplicateDrop['driver_count'].sum()
percentUrbanDriverCount = (urbanDriverCount/totalDriverCount)*100

# create lists of values needed for pie chart creation
percent = [percentUrbanDriverCount, percentSuburbanDriverCount, percentRuralDriverCount ]
labels = ['Urban', 'Suburban', 'Rural']
explode = [0.2,0,0]
colors = ['lightcoral', 'lightskyblue','gold']

# create pie chart
plt.pie(percent, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=165)

# label title of pie chart
plt.title("Percent of Total Drivers by City Type:")

# save fig to a path of choice, in this case the desktop of my computer
plt.savefig("/Users/marlonschieber/Desktop/totaldrivers.png", bbox_inches='tight', pad_inches=2)
```
